La función "boxplot" ofrece la opción de agrupar los datos por alguna variable nominal.

Por ejemplo, intentemos graficar los diagramas de caja de la variable "bmi", agrupando por el género ("gender").

```python
ax = plt.axes()
df_pacientes.boxplot(column='bmi',by='gender',figsize=(6,4),ax=ax,)
plt.title("")
plt.suptitle("")
ax.set_title('Diagrama de caja')
ax.set_ylabel('BMI')
ax.set_xlabel('Género')
```

Podemos hacer lo mismo, agrupando esta vez por la variable "hypertension", que indica si el paciente sufre hipertensión ("Yes") o no ("No").

```python
ax = plt.axes()
df_pacientes.boxplot(column='bmi',by='hypertension',figsize=(6,4),ax=ax,color='green')
plt.title("")
plt.suptitle("")
ax.set_title('Diagrama de caja')
ax.set_ylabel('BMI')
ax.set_xlabel('Sufre hipertensión')
```

Para hacer un histograma separando por el valor de una variable, tenemos que indicar el filtro en los datos.

Por ejemplo, grafiquemos la proporción de hombres y mujeres que sufren de hipertensión.

```python
# Filtro mujeres
mujeres = df_pacientes['gender'] == 'Female'
# Filtro hombres
hombres = df_pacientes['gender'] == 'Male'


plt.figure(figsize=(6,4))
plt.title("Pacientes")
###############

df_pacientes[hombres]['hypertension'].value_counts().plot(kind='bar', color='green', edgecolor='black',label='Hombres')
df_pacientes[mujeres]['hypertension'].value_counts().plot(kind='bar', color='orange', edgecolor='black',label='Mujeres')

#############
plt.legend()
plt.xlabel('Sufre hipertensión')
plt.ylabel('Frecuencia')
plt.show()
```

### ¿Qué puede concluir de los resultados?

¿Podemos decir si, en la muestra que estamos analizando, hay una tendencia a sufrir hipertensión dependiendo del género del paciente?

¡Obtengamos los números que lo sustentan!

Primero calcularemos el porcentaje de mujeres que sufren hipertensión de la muestra de datos.

```python
# Número de mujeres
nmujeres=len(df_pacientes[mujeres])

# Número de mujeres con hipertensión
nmujeres_hyp=len(df_pacientes[(mujeres)&(df_pacientes['hypertension']=='Yes')])

# Tasa
hyp_mujeres=nmujeres_hyp/nmujeres

print(f"Porcentaje de mujeres con hipertensión: {hyp_mujeres:.2%}")
```
