## Limpieza y elección de variables

Como puede notar, no todas las variables son útiles para nosotros en este momento.

```python
columnas=['Año','fecha','Clave ciudad','Precio promedio','Genérico']

df_nuevo = df[columnas]

df_nuevo.head()
```

Podemos intentar filtrar los datos, por ejemplo, para elegir una ciudad en específico.

Intentemos seleccionar los datos de la ciudad de "Puebla", y eliminemos los datos correspondientes con los precios de los "Aceites lubricantes".

```python
puebla = (df_nuevo['Clave ciudad']==???)&(df_nuevo['Genérico']!='Aceites lubricantes')

df_puebla=df_nuevo[puebla]

df_puebla.describe()
```

Note que en los datos tenemos el precio promedio en función de un índice.

En este caso, el valor 100 indica que el precio era el mismo que el promedio de la primera quincena de julio del 2018 para ese tipo de combustible.

Veamos cómo luce el precio a lo largo del tiempo para la ciudad de Puebla con una gráfica de dispersión:

```python
gasolina_bajo = df_puebla['Genérico'] == 'Gasolina de bajo octanaje'
gasolina_alto = df_puebla['Genérico'] == 'Gasolina de alto octanaje'

plt.figure(figsize=(6,4))
plt.title('Precio promedio de gasolina en Puebla')

plt.scatter(df_puebla[gasolina_bajo]['fecha'],df_puebla[gasolina_bajo]['Precio promedio'], color='blue',label='Bajo octanaje')
plt.scatter(df_puebla[gasolina_alto]['fecha'],df_puebla[gasolina_alto]['Precio promedio'], color='red',label='Alto octanaje')

plt.legend()
plt.xlabel('Fecha')
plt.ylabel('Precio promedio (índice)')
```

### Seleccionando un subconjunto de ciudades

Ahora intentemos seleccionar algunas ciudades para comparar.

A continuación se presentan dos arreglos (listas). En "ciudades" enlistaremos las ciudades con un nombre simplificado, que queremos comparar.

Luego, en el arreglo "codigos", colocar las claves en el orden correspondiente.

```python
# seleccionar algunas ciudades

# Colocar aquí los nombres simplificados
ciudades=['Puebla',???]

# Colocar aquí la clave correspondiente
codigos=[11,???]



df_ciudades=df_nuevo[(df_nuevo['Clave ciudad'].isin(codigos))&(df_nuevo['Genérico']!='Aceites lubricantes')]
df_ciudades['Ciudad']=df_ciudades['Clave ciudad'].replace({codigos[i]:str(ciudades[i]) for i in range(len(ciudades))},inplace=True)

# display(df_ciudades['Ciudad'].value_counts)
# display(df_ciudades.describe())
```

```python
# GRAFIQUEMOS EL PRECIO PROMEDIO PARA GASOLINA DE BAJO OCTANAJE

df_ciudades[df_ciudades['Genérico']=='Gasolina de bajo octanaje'].boxplot(column='Precio promedio',by='Ciudad',figsize=(10,10))
```

Una comparación un poco más elaborada puede ser ver el precio promedio de cada ciudad por Año.

```python
colors = ['red', 'blue', 'green', 'magenta', 'cyan', 'orange']

fig, ax = plt.subplots(figsize=(10,8))

# Graficar cada ciudad con un color distinto
for i in range(len(ciudades)):
    name = ciudades[i]
    mask = (df_ciudades['Ciudad']==name) & (df_ciudades['Genérico']=='Gasolina de alto octanaje')
    df_ciudades[mask].boxplot(column='Precio promedio',by='Año', ax=ax,color=colors[i],label=name)

plt.legend()
plt.xlabel('Año')
plt.ylabel('Precio promedio (índice)')
plt.show()
```
